Feature Scaling to normalise the featueres.

```python
# Import for scaling
from sklearn.preprocessing import StandardScaler

# Separate features (X) and target (y)
X = df.drop('Class', axis=1)
y = df['Class']

# Initialize scaler
scaler = StandardScaler()

# Apply scaling
X_scaled = scaler.fit_transform(X)

print("Scaling applied. Shape of X_scaled:", X_scaled.shape)
```

Train - Test - Split

```python
# Import for train-test split
from sklearn.model_selection import train_test_split

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, stratify=y, random_state=42
)

print("Training set shape:", X_train.shape)
print("Testing set shape:", X_test.shape)
print("Class distribution in Test set:")
print(y_test.value_counts())
```

Applying SMOTE on Training set only.

```python
# Import SMOTE
from imblearn.over_sampling import SMOTE

# Initialize SMOTE
smote = SMOTE(random_state=42)

# Fit and resample on training data
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

print("\nAfter SMOTE on training data:")
print(y_train_res.value_counts())
print("New training shape:", X_train_res.shape)
```

Before and After SMOTE

```python
# Import libraries for visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Class distribution BEFORE SMOTE (Original Training Set)
plt.figure(figsize=(10,4))
sns.countplot(x=y_train, palette='coolwarm')
plt.title('Class Distribution Before SMOTE')
plt.xlabel('Class (0 = Non-Fraud, 1 = Fraud)')
plt.ylabel('Count')
plt.show()

# Class distribution AFTER SMOTE (Resampled Training Set)
plt.figure(figsize=(10,4))
sns.countplot(x=y_train_res, palette='Set2')
plt.title('Class Distribution After SMOTE')
plt.xlabel('Class (0 = Non-Fraud, 1 = Fraud)')
plt.ylabel('Count')
plt.show()
```

Applying Models

```python
# Import shared libraries
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, roc_curve, auc, roc_auc_score

# Evaluation function
def evaluate_model(model_name, y_true, y_pred, y_probs):
    print(f"\n===== {model_name} =====")
    print(classification_report(y_true, y_pred, digits=4))
    print("Confusion Matrix:")
    print(confusion_matrix(y_true, y_pred))
    auc_score = roc_auc_score(y_true, y_probs)
    print(f"AUC-ROC: {auc_score:.4f}")

# Confusion Matrix Plot
def plot_confusion_matrix(model_name, y_true, y_pred):
    disp = ConfusionMatrixDisplay.from_predictions(y_true, y_pred, cmap='Blues')
    plt.title(f'Confusion Matrix: {model_name}')
    plt.show()

# ROC Curve Plot
def plot_roc_curve(model_name, y_true, y_probs):
    fpr, tpr, _ = roc_curve(y_true, y_probs)
    roc_auc = auc(fpr, tpr)
    plt.figure(figsize=(6,4))
    plt.plot(fpr, tpr, label=f'{model_name} (AUC={roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve: {model_name}')
    plt.legend(loc='lower right')
    plt.show()
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

# Train model
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train_res, y_train_res)

# Predict
y_pred_lr = lr.predict(X_test)
y_probs_lr = lr.predict_proba(X_test)[:, 1]

# Evaluate
evaluate_model("Logistic Regression", y_test, y_pred_lr, y_probs_lr)
print("\n" + "="*80 + "\n")


# Confusion Matrix
plot_confusion_matrix("Logistic Regression", y_test, y_pred_lr)
print("\n" + "="*80 + "\n")


# ROC Curve
plot_roc_curve("Logistic Regression", y_test, y_probs_lr)
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

# Train Random Forest with optimized parameters
rf = RandomForestClassifier(n_estimators=50, max_depth=15, n_jobs=-1, random_state=42)
rf.fit(X_train_res, y_train_res)

# Predict
y_pred_rf = rf.predict(X_test)
y_probs_rf = rf.predict_proba(X_test)[:, 1]

# Evaluate
evaluate_model("Random Forest", y_test, y_pred_rf, y_probs_rf)
print("\n" + "="*80 + "\n")


# Confusion Matrix
plot_confusion_matrix("Random Forest", y_test, y_pred_rf)
print("\n" + "="*80 + "\n")


# ROC Curve
plot_roc_curve("Random Forest", y_test, y_probs_rf)
```

XGBoost

```python
from xgboost import XGBClassifier

# Train XGBoost with optimized parameters
xgb = XGBClassifier(
    n_estimators=100,
    max_depth=5,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    n_jobs=-1,
    eval_metric='logloss',
    use_label_encoder=False,
    random_state=42
)

xgb.fit(X_train_res, y_train_res)

# Predict
y_pred_xgb = xgb.predict(X_test)
y_probs_xgb = xgb.predict_proba(X_test)[:, 1]

# Evaluate
evaluate_model("XGBoost", y_test, y_pred_xgb, y_probs_xgb)
print("\n" + "="*80 + "\n")

# Confusion Matrix
plot_confusion_matrix("XGBoost", y_test, y_pred_xgb)
print("\n" + "="*80 + "\n")

# ROC Curve
plot_roc_curve("XGBoost", y_test, y_probs_xgb)
```

ANN (Neural Network)

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# Build ANN
nn_model = Sequential()
nn_model.add(Dense(32, activation='relu', input_dim=X_train_res.shape[1]))
nn_model.add(Dropout(0.3))
nn_model.add(Dense(16, activation='relu'))
nn_model.add(Dense(1, activation='sigmoid'))

# Compile
nn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train (optimized for speed)
nn_model.fit(X_train_res, y_train_res, epochs=5, batch_size=128, validation_split=0.2, verbose=1)

# Predict
y_probs_nn = nn_model.predict(X_test).ravel()
y_pred_nn = (y_probs_nn > 0.5).astype(int)

# Evaluate
evaluate_model("Neural Network", y_test, y_pred_nn, y_probs_nn)
print("\n" + "="*80 + "\n")

# Confusion Matrix
plot_confusion_matrix("Neural Network", y_test, y_pred_nn)
print("\n" + "="*80 + "\n")

# ROC Curve
plot_roc_curve("Neural Network", y_test, y_probs_nn)
```

Autoencoder

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

# Define Autoencoder
encoding_dim = 14
input_dim = X_train.shape[1]

autoencoder = Sequential([
    Dense(encoding_dim, activation='relu', input_shape=(input_dim,)),
    Dense(encoding_dim // 2, activation='relu'),
    Dense(encoding_dim, activation='relu'),
    Dense(input_dim, activation='sigmoid')
])

autoencoder.compile(optimizer='adam', loss='mse')
autoencoder.fit(X_train, X_train, epochs=10, batch_size=64, validation_split=0.2)

# Reconstruction error
reconstructions = autoencoder.predict(X_test)
mse = np.mean(np.power(X_test - reconstructions, 2), axis=1)
threshold = np.percentile(mse, 95)

# Predict anomalies
y_pred_ae = [1 if e > threshold else 0 for e in mse]

# Evaluate
evaluate_model("Autoencoder", y_test, y_pred_ae, mse)
print("\n" + "="*80 + "\n")

# Confusion Matrix
plot_confusion_matrix("Autoencoder", y_test, y_pred_ae)
print("\n" + "="*80 + "\n")

# ROC Curve
plot_roc_curve("Autoencoder", y_test, mse)
```
